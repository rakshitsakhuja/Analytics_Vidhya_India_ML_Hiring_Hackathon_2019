## Applying F_Regression to choose best features

```python
from sklearn.feature_selection import f_regression,SelectKBest
X_new=SelectKBest(f_regression,k=19).fit(X,Y)
newfeatures=X.columns[X_new.get_support()]
newfeatures.values
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X=train_data[newfeatures.values]
Y=train_Target
```

## Handling Imbalanced Dataset using SMOTE and cleaning using Tomek links.

```python
from imblearn.combine import SMOTETomek

smt = SMOTETomek(ratio='auto')
X_sm, y_sm = smt.fit_sample(X, Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.33, random_state=42)
```

## Applying Random Forest Regressor

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
clf = RandomForestClassifier(n_estimators=100, max_depth=10,random_state=42)
clf.fit(X_train, y_train)  
Ytrain_Pred=clf.predict(X_train)
f1_score(y_train, Ytrain_Pred)
```

```python
clf.score(X_train, y_train)
```

```python
Ytest_Pred=clf.predict(X_test)
f1_score(y_test, Ytest_Pred)
clf.score(X_test,y_test)
```

```python
def f1_eval(y_pred, dtrain):
    y_true = dtrain.get_label()
    err = 1-f1_score(y_true, np.round(y_pred))
    return 'f1_err', err
```

```python
params = {"objective":"reg:logistic",'colsample_bytree': 0.4,'learning_rate': 0.01,
                'max_depth': 20, 'alpha': 10,'n_estimators':300,'subsample':0.3}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=10,
                    num_boost_round=50,early_stopping_rounds=50, metrics='error', seed=123)
```
